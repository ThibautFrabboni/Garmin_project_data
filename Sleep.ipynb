```python
import pandas as pd 
import json 
import datetime
import os
```

```python
#Cette fonction permet de récupérer les features choisis dans les données de sommeil de l'utilisateur.
#Elle prend en entrée le fichier json de sommeil d'un utilisateur et un nom pour le fichier csv de sortie afin d'assurer la tracabilité.
def SLEEP_data(fichier, nom : str): 
    with open(fichier) as SLEEP:
        data = json.load(SLEEP)
        data_list = []
    for entry in data: 
        #sleepStartTimestampGMT 
        if 'sleepStartTimestampGMT' in entry:
            sleepStartTimestampGMT = entry["sleepStartTimestampGMT"]
        else:
            sleepStartTimestampGMT = None
        #sleepEndTimestampGMT
        if 'sleepEndTimestampGMT' in entry:
            sleepEndTimestampGMT = entry["sleepEndTimestampGMT"]
        else:
            sleepEndTimestampGMT = None
        #calendarDate
        if 'calendarDate' in entry:
            calendarDate = entry["calendarDate"]
        else:
            calendarDate = None
        #deepSleepSeconds
        if 'deepSleepSeconds' in entry:
            deep_sleep = entry["deepSleepSeconds"]
        else:
            deep_sleep = None
        #lightSleepSeconds
        if 'lightSleepSeconds' in entry:
            light_sleep = entry["lightSleepSeconds"]
        else:
            light_sleep = None
        #remSleepSeconds sommeil paradoxal (dernière phase du cycle de sommeil)
        if 'remSleepSeconds' in entry:
            rem_sleep = entry["remSleepSeconds"]
        else:
            rem_sleep = None
        #awakeSleepSeconds
        if 'awakeSleepSeconds' in entry:
            awake_sleep = entry["awakeSleepSeconds"]
        else:
            awake_sleep = None
        #sleepScore
        if 'sleepScores' in entry and 'overallScore' in entry["sleepScores"]:
            sleepScore = entry["sleepScores"]["overallScore"]
        else:
            sleepScore = None



        #Transforme la liste en dataframe
        data_list.append({'sleepStartTimestampGMT': sleepStartTimestampGMT, 'sleepEndTimestampGMT': sleepEndTimestampGMT, 
        'Date': calendarDate, 'deep_sleep': deep_sleep, 'light_sleep': light_sleep, 'rem_sleep': rem_sleep, 'awake_sleep': awake_sleep, 
        'sleepScore': sleepScore

                        
                        })


    df_SLEEP = pd.DataFrame(data_list)
    #pré-traitement des données qui ne sera pas appliqué ici au final
    # df_SLEEP['sleepStartTimestampGMT'] = pd.to_datetime(df_SLEEP['sleepStartTimestampGMT']).dt.strftime('%Y-%m-%d %H:%M:%S')
    # df_SLEEP['sleepEndTimestampGMT'] = pd.to_datetime(df_SLEEP['sleepEndTimestampGMT']).dt.strftime('%Y-%m-%d %H:%M:%S')
    # df_SLEEP['Durée'] = pd.to_datetime(df_SLEEP['sleepEndTimestampGMT']) - pd.to_datetime(df_SLEEP['sleepStartTimestampGMT'])
    # df_SLEEP['Durée'] = df_SLEEP['Durée'].dt.total_seconds().apply(lambda x: pd.to_datetime(x, unit='s').strftime('%H:%M:%S'))

    # df_SLEEP['deep_sleep'] = df_SLEEP['deep_sleep'].fillna(0)
    # df_SLEEP['light_sleep'] = df_SLEEP['light_sleep'].fillna(0)
    # df_SLEEP['rem_sleep'] = df_SLEEP['rem_sleep'].fillna(0)
    # df_SLEEP['awake_sleep'] = df_SLEEP['awake_sleep'].fillna(0)

    #Chemin du dossier pour stocker les fichiers exportés 
    file_path = '/Users/thibaut/Desktop/Garmin_Project_ML/Data/SLEEP/'
    if not os.path.exists(file_path):
        os.makedirs(file_path)
    df_SLEEP.to_csv(f"{file_path}SLEEP_data_{nom}.csv", index=False)
    
```
