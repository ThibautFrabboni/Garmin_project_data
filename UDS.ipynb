```python
import pandas as pd 
import json 
import os
```

```python
def UDS_data(fichier, nom: str): 
    with open(fichier) as UDS: 
        data = json.load(UDS)
        data_list = []

    for entry in data:
        #calendarDate -> Date des données journalière
        if "calendarDate" in entry:
            date = entry.get("calendarDate")
        else:
            date = None
        #activeKilocalories -> nombre de calories actives dans la journée 
        if "activeKilocalories" in entry:
            active_kcal = entry.get("activeKilocalories")
        else:
            active_kcal = None
        #totalSteps -> nombre de pas dans la journée
        if "totalSteps" in entry:
            total_steps = entry.get("totalSteps")
        else:
            total_steps = None
        #totalDistanceMeters -> distance en mètre parcourue dans la journée
        if "totalDistanceMeters" in entry:
            total_distance = entry.get("totalDistanceMeters")
        else:
            total_distance = None
        #HighlyActiveSeconds -> nombre de secondes très actives (à vérifier ce que c'est) dans la journée
        if "highlyActiveSeconds" in entry:
            highly_active_seconds = entry.get("highlyActiveSeconds")
        else:
            highly_active_seconds = None
        #ActiveSeconds -> nombre de secondes actives dans la journée
        if "activeSeconds" in entry:
            active_seconds = entry.get("activeSeconds")
        else:
            active_seconds = None
        #floorsAscendedInMeters -> nombre d'étages montés dans la journée (à vérifier ce que c'est)
        if "floorsAscendedInMeters" in entry:
            floorAscended = entry.get("floorsAscendedInMeters")
        else:
            floorAscended = None
        #floorsDescendedInMeters -> nombre d'étages descendus dans la journée (à vérifier ce que c'est)
        if "floorsDescendedInMeters" in entry:
            floorDescended = entry.get("floorsDescendedInMeters")
        else:
            floorDescended = None
        #minHeartRate -> FC la plus basse de la journée
        if "minHeartRate" in entry: 
            min_heart_rate = entry.get("minHeartRate")
        else:
            min_heart_rate = None
        #maxHeartRate -> FC la plus haute de la journée
        if "maxHeartRate" in entry:
            max_heart_rate = entry.get("maxHeartRate")
        else:
            max_heart_rate = None
        #restingHeartRate -> FC au repos 
        if "restingHeartRate" in entry:
            resting_heart_rate = entry.get("restingHeartRate")
        else:
            resting_heart_rate = None
        #averageStressLevel -> niveau de stress moyen de la journée (à vérifier ce que c'est)
        if "allDayStress" in entry and "aggregatorList" in entry["allDayStress"]:
            stress_total = entry["allDayStress"]["aggregatorList"][0]["averageStressLevel"]
        else:
            stress_total = None
        #maxStressLevel -> niveau de stress max de la journée (à vérifier ce que c'est)
        if "allDayStress" in entry and "aggregatorList" in entry["allDayStress"] and "maxStressLevel" in entry["allDayStress"]["aggregatorList"][0]:
            stress_max = entry["allDayStress"]["aggregatorList"][0]["maxStressLevel"]
        else:
            stress_max = None
        #activityDuration -> durée d'action de la journée
        if "allDayStress" in entry and "aggregatorList" in entry["allDayStress"] and  "activityDuration" in entry["allDayStress"]["aggregatorList"][0]:    
            activity_duration = entry["allDayStress"]["aggregatorList"][0]["activityDuration"]
        else:
            activity_duration = None
        #bodybatteryhighest -> le plus haut niveau de body battery de la journée
        if "bodyBattery" in entry and "bodyBatteryStatList" in entry["bodyBattery"]:
            bodybattery_highest = entry["bodyBattery"]["bodyBatteryStatList"][0]["statsValue"]
        else:
            bodybattery_highest = None
        #bodybatterylowest -> le plus bas niveau de body battery de la journée
        if "bodyBattery" in entry and "bodyBatteryStatList" in entry["bodyBattery"]:
            bodybattery_lowest = entry["bodyBattery"]["bodyBatteryStatList"][1]["statsValue"]
        else:
            bodybattery_lowest = None
        #averageMonitoringEnvironmentAltitude -> altitude moyenne de la journée (à vérifier ce que c'est)
        if 'averageMonitoringEnvironmentAltitude' in entry:
            averageMonitoringEnvironmentAltitude = entry["averageMonitoringEnvironmentAltitude"]
        else:
            averageMonitoringEnvironmentAltitude = None
        #sweatLossInML -> transpiration en ml de la journée , que quand il y'a des activités sportives
        if 'hydration' in entry and 'sweatLossInML' in entry["hydration"]:
            sweat_loss = entry["hydration"]["sweatLossInML"] 
        else:
            sweat_loss = None
        #avgWakingRespirationValue = nombre de respiration par minute (rpm) qui inclu inspiration et expiration
        if 'respiration' in entry and 'avgWakingRespirationValue' in entry["respiration"]:
            respiration_value = entry["respiration"]["avgWakingRespirationValue"]
        else:
            respiration_value = None
    





        data_list.append({'Date': date, 'active_kcal':active_kcal, 'total_steps': total_steps, 'total_distance': total_distance, 
                        'highly_active_seconds': highly_active_seconds, 'active_seconds': active_seconds, 'floorAscended': floorAscended,
                            'floorDescended': floorDescended, 'min_heart_rate': min_heart_rate, 'max_heart_rate': max_heart_rate,
                            'resting_heart_rate': resting_heart_rate, 'AverageStressLevel_TOTAL': stress_total, 
                            'MaxStressLevel_TOTAL': stress_max, 'activityDuration': activity_duration, 'bodybattery_highest': bodybattery_highest,
                            'bodybattery_lowest': bodybattery_lowest, 'averageMonitoringEnvironmentAltitude': averageMonitoringEnvironmentAltitude,
                            'sweat_loss': sweat_loss, 'respiration_value': respiration_value

                            
                            
                            })

    df_UDS = pd.DataFrame(data_list)

    file_path = '/Users/thibaut/Desktop/Garmin_Project_ML/Data/UDS/'
    if not os.path.exists(file_path):
        os.makedirs(file_path)
    df_UDS.to_csv(f"{file_path}UDS_data_{nom}.csv", index=False)
    #return df_UDS
```
